## Monte Carlo estimation 

가장 흔한 예시를 들어 Monte Carlo estimation에 대해 이해해봅시다.

example : 

1 by 1 정사각형 안에 반지름이 1인 원 (1/4조각)의 넓이를 구해 원주율 $\pi$을 구해봅시다.

무작위 랜덤한 포인트 100, 1000, 10000번 추출해, 만약 원점으로부터의 거리가 1보다 가까우면 주황색 점을, 1보다 멀면 파란색 점을 찍어보도록 하겠습니다. 그리고 그 점의 비율을 토대로 원주율을 구해보겠습니다.

### 100번

```python
import numpy as np

# 원점으로부터 1보다 멀리 떨어진 점을 저장
points_outside = []

# 원점으로부터 1보다 가까운 점을 저장
points_inside = []

# 100번 시행
for _ in range(100):
    x = np.random.rand(1)
    y = np.random.rand(1)
    if np.sqrt(x**2 + y**2)>1:
        points_outside.append([x,y])
    else:
        points_inside.append([x,y])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.scatter(np.array(points_outside)[:,0],np.array(points_outside)[:,1],alpha=0.5,s=5)
plt.scatter(np.array(points_inside)[:,0],np.array(points_inside)[:,1],alpha=0.5,s=5)
```

```python
# 원주율은 다음과 같음
len(points_inside)/(len(points_inside)+len(points_outside))*4
```
